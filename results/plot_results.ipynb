```python
#Libraries
import matplotlib.pyplot as plt
import xlrd 
```

```python
results_path = "results.xlsx"
wb = xlrd.open_workbook(results_path) 
sheet = wb.sheet_by_index(0) 

num_particles = sheet.col_values(0)
error_x	= sheet.col_values(1)
error_y	= sheet.col_values(2)
error_yaw = sheet.col_values(3)
system_time = sheet.col_values(4)

# plot
f, axs = plt.subplots(2, 2, figsize=(24, 9), dpi=80, facecolor='w', edgecolor='k')
# f.suptitle('Error plot for different particle sizes', fontsize=30)

axs[0,0].set_title('error_x', fontsize=25)
axs[0,1].set_title('error_y', fontsize=25)
axs[1,0].set_title('error_yaw', fontsize=25)
axs[1,1].set_title('system_time', fontsize=25)

axs[0,0].set_xlabel('number of particles', fontsize=10)
axs[0,1].set_xlabel('number of particles', fontsize=10)
axs[1,0].set_xlabel('number of particles', fontsize=10)
axs[1,1].set_xlabel('number of particles', fontsize=10)

axs[0,0].plot(num_particles[1:], error_x[1:],'-o',linewidth=4)
axs[0,1].plot(num_particles[1:], error_y[1:],'-o',linewidth=4)
axs[1,0].plot(num_particles[1:], error_yaw[1:],'-o',linewidth=4)
axs[1,1].plot(num_particles[1:], system_time[1:],'-o',linewidth=4)


plt.subplots_adjust(left=0., right=1.5, top=1.4, bottom =0)
plt.show()
f.savefig("error_characteristics.jpg", bbox_inches = 'tight')
```
